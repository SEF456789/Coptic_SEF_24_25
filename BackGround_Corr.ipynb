```python
from pydub import AudioSegment
import librosa
import soundfile as sf
import numpy as np
import os

input_folder='Audio_Temp'
output_folder='Audios'

# Audio amplification factor (1.0 = no change, 2.0 = double volume, 0.5 = half volume)
amplification_factor = 2.5  # Increase volume by 150%

for filename in os.listdir(input_folder):
    if filename.endswith('.opus'):
        input_path = os.path.join(input_folder, filename)
        output_path = os.path.join(output_folder, filename.replace('.opus', '.wav'))
        
        # Convert .opus to .wav
        audio = AudioSegment.from_file(input_path)
        audio.export(output_path, format="wav")
        
        # Load audio for processing
        y, sr = librosa.load(output_path, sr=48000)
        
        # Apply amplification/magnification
        amplified_audio = y * amplification_factor
        
        # Prevent clipping by normalizing if amplitude exceeds [-1, 1]
        max_amplitude = np.max(np.abs(amplified_audio))
        if max_amplitude > 1.0:
            amplified_audio = amplified_audio / max_amplitude
            print(f"Audio clipped, normalized from {max_amplitude:.2f} to 1.0 for {filename}")
        
        # Save the processed audio
        sf.write(output_path, amplified_audio, sr)
        print(f"Converted and amplified {filename} to {output_path} (factor: {amplification_factor}x)")
    else:
        print(f"Skipped {filename}, not an .opus file")
```
